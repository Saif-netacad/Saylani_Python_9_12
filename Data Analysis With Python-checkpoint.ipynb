# Plotting with Matplotlib using Pandas

```python
df_grouped = df.groupby("Segment")["Sales"].mean().reset_index()

plt.figure(figsize=(10, 5))  # Adjust figure size as needed
plt.plot(df_grouped["Segment"], df_grouped["Sales"], marker="o", linestyle="-")
plt.xlabel("Segment")
plt.ylabel("Average Sales")
plt.title("Average Sales by Segment")
plt.grid(True)
plt.show()
```

```python
plt.bar(df["Segment"], df["Sales"], color='blue')  # Adjust color as needed
plt.xlabel("Segment")
plt.ylabel("Sales")  # Adjust label if applicable
plt.title("Sales by Segment (Bar Chart)")
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability with many categories
plt.grid(axis='y')  # Display grid lines on the y-axis
plt.tight_layout()  # Adjust spacing between elements for readability
plt.show()
```
